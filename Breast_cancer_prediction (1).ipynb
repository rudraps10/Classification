```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/breast_cancer_prediction.csv')
df.head()
```

```python
df=df.drop(columns=['Unnamed: 32','id'])
```

```python
for col in df.columns:
  plt.figure(figsize=(9,6))
  sns.boxplot(data=df,x=col)
```

```python
df.sample()
```

```python
for i in df.select_dtypes(include='number').columns:
  q3=df[i].quantile(0.75)
  q1=df[i].quantile(0.25)
  iqr=q3-q1
  lowlim=q1-1.5*iqr
  uplim=q3+1.5*iqr
  df[i]=np.where(df[i]<lowlim, lowlim, np.where(df[i]>uplim,uplim,df[i]))
```

```python
for i in df.select_dtypes(include='number').columns:
  plt.figure(figsize=(6,4))
  sns.boxplot(data=df,x=i)
```

```python
for i in df.select_dtypes(include='number').columns:
  plt.figure(figsize=(6,4))
  sns.displot(data=df,x=i,kde='true')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df['diagnosis']=df['diagnosis'].str.replace('M','1')
df['diagnosis']=df['diagnosis'].str.replace('B','0')
df['diagnosis'].value_counts()
```

```python
df['diagnosis'].value_counts().plot(kind='bar')
```

```python
sizes = df['diagnosis'].value_counts()
plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%', colors=['#66b3ff','#ff9999'], wedgeprops=dict(width=0.4))
plt.show()
```

```python
df["diagnosis"] = df["diagnosis"].astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop('diagnosis',axis=1),df['diagnosis'],test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression()
clf.fit(X_train_scaled,y_train)
```

```python
y_pred=clf.predict(X_test_scaled)
```

```python
df.info()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy score : ", accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
print("Precision score : ", precision_score(y_test,y_pred))
print("Recall score : ", recall_score(y_test,y_pred))
print("F1 score : ", f1_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=['Pred_0','Pred_1'],
            yticklabels=['Actual_0','Actual_1'])
plt.show()
```
